```python
# Import medium English model
import spacy
nlp = spacy.load('en_coref_md')
```

```python
# Read in data file
import pandas as pd
data = pd.read_csv('data.tsv', sep='\t')
data.set_index(data.columns[0], inplace=True)

data.head()
```

```python
# Perform coreference resolution on each sentence, and replace the pronouns with the resolved entities
# Print the new sentences, and add them to a dataframe

resolved_text = pd.DataFrame()
for row in data.iterrows():
    resolved_text = resolved_text.append([row[1]['Text']], ignore_index=True)
    print(nlp(row[1]['Text'])._.coref_resolved)

# resolved_text.head()
```
