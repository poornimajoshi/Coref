```python
import pandas as pd
```

```python
df = pd.read_csv("resolved_words.csv")
df.head()
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df['ID'] = df['0'].str.extract('(development-[0-9]{1,4})', expand=True)
```

```python
df['Resolution'] = df['0']
```

```python
df['Resolution'] = df['Resolution'].str.replace('\d+', '')
```

```python
data = df
```

```python
data=data.drop(columns=['0'])
```

```python
data['Resolution'] = data['Resolution'].map(lambda x: x.lstrip('([0-9])'))
```

```python
data.head()
```

```python
data.to_csv('core_nlp_resolved.csv')
```
